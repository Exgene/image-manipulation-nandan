```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('testImage.png', cv2.IMREAD_GRAYSCALE)
```

```python
# cv2.imshow('image', img)
# cv2.waitKey(0)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
img
```

```python
planes = np.array([cv2.bitwise_and(img, 1 << i) for i in range(7, -1, -1)])
```

```python
for i in planes:
    plt.imshow(i, cmap='gray')
    plt.title('Plane ' + str(i))
    plt.axis('off')
    plt.show()
```
